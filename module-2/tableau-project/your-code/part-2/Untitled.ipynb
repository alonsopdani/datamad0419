```python
from pymongo import MongoClient
import pandas as pd
client = MongoClient('localhost', 27017)
db = client['companies']
collection = db.companies
```

```python
sectors = ['web', 'software', 'games_video', 'social']
```

```python
companies = collection.find({'category_code': 'web'}).limit(20)
for e in companies:
    print(e['name'])
```

```python
companies = collection.find(
    {'$and':[
        {'$or':[
            {'category_code': 'web'},
            {'category_code': 'software'},
            {'category_code': 'games_video'},
            {'category_code': 'social'}]},
        {'$or':[
            {'number_of_employees': {'$lte': 50}},
            {'number_of_employees': {'$gte': 250}}]},
#        {'offices.0.latitude': {'$ne': None}},
#        {'offices.0.latitude': {'$ne': None}},
        {'offices': { '$ne': [] }},
        {'founded_year': {'$gte': 2005}}]},
    {'name' : 1, 'category_code' : 1, 'number_of_employees' : 1,
     'founded_year' : 1, 'offices.0.latitude' : 1, 'offices.0.longitude' : 1, '_id' : 0})

print(list(companies))
```
